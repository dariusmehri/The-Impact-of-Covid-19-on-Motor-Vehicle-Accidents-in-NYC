```python
import pandas as pd
import numpy as np

path = "C:\\Users\\dmehri\\Documents\\DATA\\DOT\\"

df = pd.read_csv(path + "Motor_Vehicle_Collisions_-_Crashes.csv", error_bad_lines = False)


```

```python

df = df[["CRASH DATE", "CRASH TIME",'BOROUGH', 'ZIP CODE', 'NUMBER OF PERSONS INJURED',
       'NUMBER OF PERSONS KILLED', 'NUMBER OF PEDESTRIANS INJURED',
       'NUMBER OF PEDESTRIANS KILLED', 'NUMBER OF CYCLIST INJURED',
       'NUMBER OF CYCLIST KILLED', 'NUMBER OF MOTORIST INJURED',
       'NUMBER OF MOTORIST KILLED']]
```

Motor Crashes by Zip Code

```python
df = df.dropna()

df["CRASH DATE"] = df["CRASH DATE"].astype(str)
df["Year"] = df["CRASH DATE"].str.split("/").str[2].map(str.strip)

df["ZIP CODE"] = df["ZIP CODE"].astype(str)
df["ZIP CODE"] = df["ZIP CODE"].str.split(".").str[0].map(str.strip)
```

2019

```python
df2019 = df[df["Year"] == "2019"].reset_index(drop=True)

df2019["Number Persons Injured or Killed 2019"] = df2019["NUMBER OF PERSONS INJURED"] + df2019["NUMBER OF PERSONS KILLED"]

df2019["Number Pedestrians and Cyclists Injured or Killed 2019"] = df2019["NUMBER OF PEDESTRIANS INJURED"] + df2019["NUMBER OF PEDESTRIANS KILLED"] + df2019["NUMBER OF CYCLIST INJURED"] + df2019["NUMBER OF CYCLIST KILLED"]


df2019 = df2019[["ZIP CODE", "Number Persons Injured or Killed 2019", "Number Pedestrians and Cyclists Injured or Killed 2019"]]


df2019 = df2019.groupby(['ZIP CODE']).sum()
df2019 = df2019.add_suffix('').reset_index()

df2019 = df2019.sort_values(by = 'Number Persons Injured or Killed 2019', ascending=False).reset_index(drop=True)


df2019[0:20]
```

2020

```python
df2020 = df[df["Year"] == "2020"].reset_index(drop=True)

df2020["Number Persons Injured or Killed 2020"] = df2020["NUMBER OF PERSONS INJURED"] + df2020["NUMBER OF PERSONS KILLED"]

df2020["Number Pedestrians and Cyclists Injured or Killed 2020"] = df2020["NUMBER OF PEDESTRIANS INJURED"] + df2020["NUMBER OF PEDESTRIANS KILLED"] + df2020["NUMBER OF CYCLIST INJURED"] + df2020["NUMBER OF CYCLIST KILLED"]


df2020 = df2020[["ZIP CODE", "Number Persons Injured or Killed 2020", "Number Pedestrians and Cyclists Injured or Killed 2020"]]


df2020 = df2020.groupby(['ZIP CODE']).sum()
df2020 = df2020.add_suffix('').reset_index()

df2020 = df2020.sort_values(by = 'Number Persons Injured or Killed 2020', ascending=False).reset_index(drop=True)


df2020[0:20]
```

```python
s1 = pd.merge(df2019, df2020, how='left', on=['ZIP CODE'])
s1 = s1.fillna(0)

s1 = s1.rename(columns={'Number Persons Injured or Killed 2019': 'Persons2019', 
                        'Number Pedestrians and Cyclists Injured or Killed 2019': 'PedCycl2019',
                       'Number Persons Injured or Killed 2020':'Persons2020',
                       'Number Pedestrians and Cyclists Injured or Killed 2020':'PedCycl2020',
                       'ZIP CODE':'ZIPCODE'})


```

Convert Zip Codes to Neighboorhoods

```python
h = "C:\\Users\\dmehri\\Documents\\DATA\\DOT\\"

file1 = open(path + 'Zip Codes by Hood Raw.txt', 'r')
Lines = file1.readlines()


all_data = pd.DataFrame()


 
count = 0
# Strips the newline character
for line in Lines:
    count += 1
    #print(line)
    
    a = line
    a = a.split(",")
    print (a)
    
    qn = pd.DataFrame(columns=('Hood', 'ZIPCODE'))

    for i in range(1, len(a)):
        #print (a[0], a[i])
        
        qn.loc[i] = [a[0],a[i]]


        
    all_data = all_data.append(qn,ignore_index=False)
    
    
all_data = all_data.reset_index(drop=True)
```
